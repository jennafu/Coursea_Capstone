### 1. Importing from Wikipedia Page

```python
# import relevant packages
import pandas as pd
import matplotlib.cm as cm
import matplotlib.colors as colors
from sklearn.cluster import KMeans
!pip install folium
import folium
```

```python
# Store the data into a panda dataframe
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
list_df = pd.read_html(url, header = 0)
df = list_df[0]

# Remove rows with Borough not assigned
df = df[(df.Borough != 'Not assigned')]

# There seems to be changes in the Wikipedia page, with unique boroughs and there does not seem to be Not assigned neighborhood
df.head()
```

```python
# Shape of dataframe
df.shape
```

### 2. Importing Geospatial Data

```python
# Download geospatial data
url = 'http://cocl.us/Geospatial_data'
df2 = pd.read_csv(url)
df2.head()
```

```python
df_merge = pd.merge(df, df2, left_on='Postal code', right_on='Postal Code')
df_merge = df_merge.drop(["Postal Code"],axis=1)
df_merge.head()
```
